```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pystan
%matplotlib inline
```

```python
df1 = pd.read_csv("input/data-category.txt")
df1.head()
```

```python
df1 = df1.assign(
    Age=lambda df:df.Age/100,
    Income=lambda df:df.Income/1000
)
X = df1.iloc[:,:3].values
K = len(df1.Y.unique())
data = {"N":len(df1), "D":X.shape[1], "K":K, "X":X, "Y":df1.Y}
```

```python
fit = pystan.stan(file="stan/model10-2.stan",data=data, seed=1234)
```
